# Graphs

## Library Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Movie Graphs

### Graph 1: Movie Genre Distribution

```python
df = pd.read_csv('genres_clean.csv', engine='python')
```

```python
df.head()
```

```python
genre_counts = df['genre_name'].value_counts()

plt.figure(figsize=(10, 5))
genre_counts.plot(kind='bar')
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.title("Movie Genre Distribution", fontsize=14, fontweight="bold")
plt.xticks(rotation=90)

plt.show()
```

### Graph 2: Top 10 Production Companies

```python
df = pd.read_csv('production_companies_clean.csv', engine='python')
```

```python
df.head()
```

```python
company_counts = df['company_name'].value_counts()
top_10_companies = company_counts.head(10)

plt.figure(figsize=(10, 5))
top_10_companies.plot(kind='bar')
plt.xlabel("Production Company", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.title("Top 10 Production Companies with Most Movies", fontsize=14, fontweight="bold")
plt.xticks(rotation=90)

plt.show()
```

## TV Show Graphs

### Graphs 2 and 3: First and Latest Released Date

```python
df = pd.read_csv('TV_Show(Airings)_final_cleaned.csv', engine='python')
```

```python
df.head()
```

```python
df['First_Aired_Date'] = pd.to_datetime(df['First_Aired_Date'])
df['Latest_Aired_Date'] = pd.to_datetime(df['Latest_Aired_Date'])
```

```python
df_air = df
df_air['year'] = df_air['First_Aired_Date'].dt.year
year_counts = df_air['year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
year_counts.plot(kind='line', marker='o', linewidth=2, markersize=6)

plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of TV Shows", fontsize=12)
plt.title("TV Shows Started per Year", fontsize=14, fontweight="bold")
plt.xticks(rotation=90)

plt.show()
```

```python
df_lat = df
df_lat['year'] = df_air['Latest_Aired_Date'].dt.year
year_counts = df_lat['year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
year_counts.plot(kind='line', marker='o', linewidth=2, markersize=6)

plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of TV Shows", fontsize=12)
plt.title("TV Shows' Latest Year Release", fontsize=14, fontweight="bold")
plt.xticks(rotation=90)

plt.show()
```
